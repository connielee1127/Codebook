## The Invest India Dataset

```python
invest = pd.read_csv('investIndia.csv')
invest.head()
```

```python
invest.shape
```

Here is the Invest India dataset. There are 34 entries in this dataset with 7 columns.

```python
cols_invest = invest.columns.values.astype('str')
col_inv_list = cols_invest.tolist()
col_inv_list
```

```python
invest.value_counts(['Name']).sum()
```

Each entry in the dataset describes a unique organization; hence we do not have multiple entries of the same organization.

```python
invest.Description.str.len().max()
```

The longest description is 1201 characters!

```python
invest.value_counts(['State'])
```

Evidently, many of the NGOs in this dataset seem to be active in multiple states of India. Some of the descriptions are vague, as the names don't necessarily seem to refer to just states, but reference regions of India. For example, there are three NGOs that describe their state as 'PAN India', which means that the organization is active in the entirety of India. Another example of potentially confusing naming is found with 'Villages across UP', or 'UP', which both refer to the state of Uttar Pradesh.

```python
invest.loc[2, 'State'] = 'Delhi NCR'
invest.loc[11, 'State'] = 'Delhi NCR, Haryana'
invest.loc[12, 'State'] = 'Telangana, Andhra Pradesh'
invest.loc[18, 'State'] = 'Mumbai, Varanasi, Mira Bhayander'
invest.loc[23, 'State'] = 'Bangalore, Chennai, Hyderabad, Mumbai, Noida' 
invest.loc[25, 'State'] = 'Sitapur, Delhi NCR, Lucknow, Uttar Pradesh'
invest.loc[27, 'State'] = 'Mumbai, Delhi, Tamil Nadu, Gujarat, Uttar Pradesh, West Bengal, Rajasthan, Karnataka, Andhra Pradesh, Haryana, Madhya Pradesh' 
invest.loc[28, 'State'] = 'Delhi NCR, Jammu and Kashmir, Uttarakhand, Uttar Pradesh, Andhra Pradesh, Orissa' 
invest.loc[29, 'State'] = 'Uttar Pradesh, Uttarakhand, Hazaribagh, West Singhbhum, Jharkhand, Rajasthan, Bihar'
invest = invest.rename(columns={"State": "Location"})
invest
```

Noting that many of the 'states' under the states column were not actually states, but a mix of states, cities, and urban territories, I renamed the 'State' column to 'Location', a more accurate label. I also altered some of the values in the new 'Location' column such that the format for all of the entries were uniform: the location name followed by a singular comma.

```python
invest['Contact']
```

Exploring the links under the Contact column, there were more actual websites when compared to the Give India dataset, but there were also missing or error-inducing links.

```python
i_email = invest[invest['Email'] == 'None'].shape[0]
i_pn = invest[invest['Phone'] == 'None'].shape[0]
print('In the Invest India dataset, there are', i_email, 'missing email addresses and', i_pn, 'missing phone numbers.')
```

It seems that almost a third of the email addresses are missing, and all but one of the phone numbers is missing. This indicates that I should either collect this missing data, or drop both columns if said data is unattainable.

```python
invest['Other']
```

Looking at the Other column, besides the fact that most of the values are missing, I have no idea what these values are supposed to mean. Some look like phone numbers, others the names of individuals, still others payment information...but these values are not unified in type nor really insightful; hence I drop this column.

```python
invest = invest.drop(['Other'], axis=1)
invest
```
